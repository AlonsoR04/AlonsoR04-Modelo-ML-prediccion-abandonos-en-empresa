## *__3. ¿Cómo se va a realizar la evaluación?__*

Para realizar la evaluación de nuestro modelo vamos a seguir una serie de pasos. La evaluación estará divida en inner y outer.En la inner-evaluation se elegirá el mejor modelo clasificador con el mejor método de escalado e imputación de los datos y el ajuste de los hiperparámetros del modelo. Y en la outer-evaluation, se estimará el rendimiento a futuro del modelo. 
Es importante dividir nuestro dataset en datos de entrenamiento (usados en inner) y de test (usados en outer) para poder entrenar el modelo con una proporcion de los datos y evaluarlo con otra. En nuestro caso usaremos un Holdout de (2/3) Train y (1/3) Test. Importante recalcar que al tratarse de un dataset con clases desbalanceadas el accuracy no es una métrica fiable; y por tanto optaremos por evaluar basandonos en el balanced_accuracy.

- Inner Evaluation

    - Primero, determinamos el mejor scaler e imputer. Para ello, compararemos el balanced_accuracy medio, obtenido al hacer la cross-validation (3 k-folds) con los hiperparámetros por defecto para KNN, con el obtenido al variar el scaler (Standard, MinMax y Robust) y el imputer (usar la media o la mediana).

    - Una vez obtenido el mejor scaler e imputer, procederemos a buscar el mejor modelo. Para ello compararemos el balanced_accuracy de distintos modelos como KNN, Trees, modelos lineales y Support Vector Machines. Cada uno se comprueba con los hiper-parámetros default y con los hiper-parámetros optimizados.

    - Con el mejor modelo elegido, se realizará Outer Evaluation.

- Outer Evaluation

    - Con el mejor modelo obtenido en la fase de evaluación interna, validaremos su entrenamiento haciendo uso de la partición de datos TEST (1/3). Esto nos permitirá estimar el rendimiento del modelo elegido con vista a futuro.

Además, una vez realizadas ambas evaluaciones, someteremos a nuestro modelo a unos datos de competición que nos devolverán ciertas predicciones. A partir de esa predicción se comprobara que la partición de las distintas clases de la etiqueta tiene una distribución similar que en los datos de entrenamiento.

Tras obtener los datos del modelo Dummy, establecemos como requisito minimo que nuestros modelos tengan que superar el 51% en balanced_accuracy y el 75% en accuracy. En caso contrario, serán inmediatamente descartados.

### *__4.1 - K-nearest neigbors__*


Iterar entre los diversos metodos de escalado e imputación para ver cual obtiene mejores métricas  

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler, MinMaxScaler, RobustScaler
from sklearn.impute import SimpleImputer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import StratifiedKFold, cross_val_score, cross_val_predict, cross_validate
from sklearn.metrics import confusion_matrix, make_scorer, accuracy_score, balanced_accuracy_score

#Separar variables categoricas y numéricas
columnas_num = X.select_dtypes(include=['float64', 'int64']).columns.tolist()
columnas_cat = X.select_dtypes(include=['object']).columns.tolist()


#Distintos métodos de escalado e imputación
scalers = [StandardScaler(), MinMaxScaler(), RobustScaler()]
imputers = ['mean', 'median']

#Se realiza una crossvalidation estratificada con 3 folds
cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=seed)

accuracy = -np.inf
best_scaler, best_imputer, best_time_def = None, None, None

# Definir los pasos en la Pipeline
knn = KNeighborsClassifier()
encoder = OneHotEncoder()

# Pipeline para valores categóricos
classif_categoricos = Pipeline([
            ("encoder", encoder),
            ("imputation", SimpleImputer(strategy='most_frequent'))
        ])

# Definir funciones para TPR y TNR
def true_positive_rate(y_true, y_pred):
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    return tp / (tp + fn) if (tp + fn) > 0 else 0  # Evitar división por cero

def true_negative_rate(y_true, y_pred):
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    return tn / (tn + fp) if (tn + fp) > 0 else 0  # Evitar división por cero

# Definir las métricas de evaluación
scoring = {
    'balanced_accuracy': make_scorer(balanced_accuracy_score),
    'accuracy': make_scorer(accuracy_score),
    'tpr': make_scorer(true_positive_rate),
    'tnr': make_scorer(true_negative_rate)
}

for i in range(len(imputers)):
    for j in range(len(scalers)):
        scaler = scalers[j]
        imputer = SimpleImputer(strategy=imputers[i])

        classif_numericos = Pipeline([
            ("imputation", imputer),
            ("standardization", scaler)
        ])

        preprocessor = ColumnTransformer(
            transformers=[
                ("num", classif_numericos, columnas_num),
                ("cat", classif_categoricos, columnas_cat)
            ]
        )

        clf = Pipeline([("preprocessor", preprocessor), ("classifier", knn)])

        scr = cross_validate(clf, X_train, y_train, scoring=scoring, cv=cv, n_jobs=-1)

        mean_balanced_accuracy = scr["test_balanced_accuracy"].mean()
        mean_accuracy = scr["test_accuracy"].mean()
        mean_tpr = scr["test_tpr"].mean()
        mean_tnr = scr["test_tnr"].mean()
        time_i = scr["fit_time"].mean()

        print(f"Scaler: {scalers[j]}, Imputer: {imputers[i]}")
        print(f"  Mean Balanced Accuracy = {mean_balanced_accuracy}")
        print(f"  Mean Accuracy          = {mean_accuracy}")
        print(f"  Mean TPR               = {mean_tpr}")
        print(f"  Mean TNR               = {mean_tnr}")
        print(f"  Training Time          = {time_i:.4f} sec\n")

        if mean_balanced_accuracy > accuracy:
            accuracy = mean_balanced_accuracy
            best_scaler, best_imputer = scalers[j], imputers[i]
            best_time_def = time_i

print(f"Best balanced accuracy: {accuracy}. With scaler: {best_scaler}, imputer: {best_imputer}, trained on: {best_time_def} seconds.")
```

El mejor scaler es el MinMaxScaler, y el mejor metodo de imputación es usar la media. Estos se usarán a lo largo del desarrollo de la práctica

Una vez sabemos cual es el mejor scaler y el mejor imputer, elegimos los mejores hiperparámetros de KNN haciendo un GridSearch con cross-validation, variando los hiperparámetros número de vecinos (n_neighbors) y la metrica utilizada.

```python
from sklearn.model_selection import GridSearchCV
import time

classif_numericos = Pipeline([
    ("imputation", SimpleImputer(strategy=best_imputer)),
    ("standardization", best_scaler)
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", classif_numericos, columnas_num),
        ("cat", classif_categoricos, columnas_cat)
    ]
)

clf = Pipeline([("preprocessor", preprocessor), ("classifier", knn)])

knn_grid = {"classifier__n_neighbors": [1, 2, 3, 4, 8, 16, 32],
            "classifier__metric": ["euclidean", "manhattan"]}

grid = GridSearchCV(clf, knn_grid, cv=cv, scoring=scoring, refit='balanced_accuracy', n_jobs=-1)

grid.fit(X_train, y_train)
best_model_knn = grid.best_estimator_

start_time = time.time()
best_model_knn.fit(X_train, y_train)  # Reentrenar solo con los mejores hiperparámetros
best_fit_time = time.time() - start_time
# Extraer resultados
cv_results = grid.cv_results_

mean_acc = np.max(cv_results["mean_test_accuracy"])  # Mejor Accuracy
mean_tpr = np.max(cv_results["mean_test_tpr"])  # Mejor TPR
mean_tnr = np.max(cv_results["mean_test_tnr"])  # Mejor TNR

print(f"KNN con HPO:")
print(f"  Mean Balanced Accuracy = {grid.best_score_}")
print(f"  Mean Accuracy          = {mean_acc}")
print(f"  Mean TPR               = {mean_tpr}")
print(f"  Mean TNR               = {mean_tnr}")
print(f"  Training Time          = {best_fit_time:.4f} sec\n")
print("Mejores hiperparámetros:", grid.best_params_)
```

Los hiperparámetros optimizados obtenidos son que el valor de K (n-neighbors) sea 1 y se utilize la distancia de Manhattan (p=1).

Se va a generar un plot para ver el efecto del cambio del hiperparámetro n-neighbors y la métrica

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Crear un gráfico para cada métrica
fig, axes = plt.subplots(2, 2, figsize=(15, 10))  # 2 filas, 2 columnas

# Definir métricas y etiquetas
metrics = ["mean_test_balanced_accuracy", "mean_test_accuracy", "mean_test_tpr", "mean_test_tnr"]
titles = ["Balanced Accuracy", "Accuracy", "True Positive Rate (TPR)", "True Negative Rate (TNR)"]

# Graficar cada métrica en un subplot
for ax, metric, title in zip(axes.flatten(), metrics, titles):
    sns.lineplot(
        data=cv_results,
        x="param_classifier__n_neighbors",
        y=metric,
        hue="param_classifier__metric",  # Diferenciar por criterio
        marker="o",
        ax=ax
    )
    ax.set_title(title)
    ax.set_xlabel("N-Neighbors")
    ax.set_ylabel("Valor de la Métrica")
    ax.grid()

plt.suptitle("Efecto de N-Neighbors y criterion en las métricas del modelo", fontsize=16)
plt.tight_layout()
plt.show()
```

Se observa que cuantos menos neighbors haya, mayor será el balanced accuracy (siendo el valor con mayor balanced_accuracy n_neighbors=1) y mayor será tambien el TPR y el accuracy.   
Esto puede suponer que haya overfitting en el modelo porque se esta sobreadaptando al ruido de los datos train. Pero como la evaluacion se ha realizado con un cross-validation de 3 folds, suponemos que esto no ocurre.

El cambio de la metrica (elegir entre distancia euclída o distancia de Manhattan) no parece variar mucho pero es ligeramente mejor la distancia de Manhattan.

### *__5.1 - Modelos lineales__*

Vamos a crear tres modelos lineales basados en la funcion Logisticregresion. Usando regularización Lasso (l1), Ridge (l2), o ninguna.

Se observa que los valores de las distintas métricas de precisión son muy similares entre los tres modelos, pero son mejores sin regularización

Optimizacion de parámetros sin regularización

```python
clf = Pipeline([("preprocessor", preprocessor), ("classifier", linear_model.LogisticRegression(penalty=None, random_state=seed))])

lr_grid = {
    'classifier__solver': ['lbfgs', 'saga', 'newton-cg'],
    'classifier__tol': [1e-4, 1e-5, 1e-6],
    'classifier__max_iter': [100, 500, 1000],
    'classifier__class_weight': [None, 'balanced']
}

lr_none_grid = GridSearchCV(clf, lr_grid, cv=cv, scoring=scoring, refit='balanced_accuracy', n_jobs=-1)

lr_none_grid.fit(X_train, y_train)
best_model_lr_none = lr_none_grid.best_estimator_

start_time = time.time()
best_model_lr_none.fit(X_train, y_train)  # Reentrenar solo con los mejores hiperparámetros
best_fit_time = time.time() - start_time
# Extraer resultados
cv_results = lr_none_grid.cv_results_

mean_acc = np.max(cv_results["mean_test_accuracy"])  # Mejor Accuracy
mean_tpr = np.max(cv_results["mean_test_tpr"])  # Mejor TPR
mean_tnr = np.max(cv_results["mean_test_tnr"])  # Mejor TNR

print(f"LogisticRegression without penalty with HPO:")
print(f"  Mean Balanced Accuracy = {lr_none_grid.best_score_}")
print(f"  Mean Accuracy          = {mean_acc}")
print(f"  Mean TPR               = {mean_tpr}")
print(f"  Mean TNR               = {mean_tnr}")
print(f"  Training Time          = {best_fit_time:.4f} sec\n")
print("Mejores hiperparámetros:", lr_none_grid.best_params_)
```

Optimización de parámetros con regularización Lasso

```python
clf = Pipeline([("preprocessor", preprocessor), ("classifier", linear_model.LogisticRegression(penalty="l1", random_state=seed))])

lr_grid = {
    'classifier__solver': ['liblinear', 'saga'],  # Solvers compatibles con L1
    'classifier__C': [0.01, 0.1, 1, 10],  # Regularización
    'classifier__tol': [1e-4, 1e-5, 1e-6],  # Precisión del solver
    'classifier__class_weight': [None, 'balanced']  # Manejo de clases desbalanceadas
}

lr_l1_grid = GridSearchCV(clf, lr_grid, cv=cv, scoring=scoring, refit='balanced_accuracy', n_jobs=-1)

lr_l1_grid.fit(X_train, y_train)
best_model_lr_l1 = lr_l1_grid.best_estimator_

start_time = time.time()
best_model_lr_l1.fit(X_train, y_train)  # Reentrenar solo con los mejores hiperparámetros
best_fit_time = time.time() - start_time
# Extraer resultados
cv_results = lr_l1_grid.cv_results_

mean_acc = np.max(cv_results["mean_test_accuracy"])  # Mejor Accuracy
mean_tpr = np.max(cv_results["mean_test_tpr"])  # Mejor TPR
mean_tnr = np.max(cv_results["mean_test_tnr"])  # Mejor TNR

print(f"LogisticRegression with Lasso penalty with HPO:")
print(f"  Mean Balanced Accuracy = {lr_l1_grid.best_score_}")
print(f"  Mean Accuracy          = {mean_acc}")
print(f"  Mean TPR               = {mean_tpr}")
print(f"  Mean TNR               = {mean_tnr}")
print(f"  Training Time          = {best_fit_time:.4f} sec\n")
print("Mejores hiperparámetros:", lr_l1_grid.best_params_)
```

Al contrario de cuando se han evaluado los modelos por omisión de hiper-parámetros. Ahora tras optimizar los hiper-parámetros es mejor el modelo con regularización Lasso antes que el que no tiene regularización

```python
# Métricas obtenidas para ambos modelos
metrics = ['Balanced Accuracy', 'Accuracy', 'TPR', 'TNR']
lr_none_scores = [
    lr_none_grid.best_score_,
    np.max(lr_none_grid.cv_results_["mean_test_accuracy"]),
    np.max(lr_none_grid.cv_results_["mean_test_tpr"]),
    np.max(lr_none_grid.cv_results_["mean_test_tnr"]),
]

lr_l1_scores = [
    lr_l1_grid.best_score_,
    np.max(lr_l1_grid.cv_results_["mean_test_accuracy"]),
    np.max(lr_l1_grid.cv_results_["mean_test_tpr"]),
    np.max(lr_l1_grid.cv_results_["mean_test_tnr"]),
]

# Crear DataFrame para facilitar el plotting
df_plot = pd.DataFrame({
    'Métrica': metrics * 2,
    'Valor': lr_none_scores + lr_l1_scores,
    'Modelo': ['LogReg (sin penalización)'] * 4 + ['LogReg (L1 - Lasso)'] * 4
})

# Estilo del gráfico
sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=df_plot, x='Métrica', y='Valor', hue='Modelo', palette='viridis')

# Anotar los valores en las barras
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', label_type='edge', padding=3)

plt.title("Comparación de Regresión Logística (Sin Penalización vs. L1)")
plt.ylim(0, 1.05)
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
```

### *__5.2 - SVMs__*

Vamos a realizar un classifier basado en Support Vector Machines (SVM), para ello usaremos SVC. Probaremos la versión sin y con HPO. Como se observa en el resultado, hay una gran diferencia entre el balanced accuracy de la version default y la optimizada pero a su vez se observa que el tiempo de entrenamiento es del doble. Aun así consideramos que es un tradeoff asequible ya que aumentamos el balanced_accuracy en un 23%.

```python
from sklearn import svm


clf = Pipeline([("preprocessor", preprocessor), ("classifier", svm.SVC(random_state=seed))]) # Definir el clasificador SVC

scr = cross_validate(clf, X_train, y_train, scoring=scoring, cv=cv) # evaluacion sin HPO

# Metricas de evaluación
mean_balanced_accuracy = scr["test_balanced_accuracy"].mean()
mean_accuracy = scr["test_accuracy"].mean()
mean_tpr = scr["test_tpr"].mean()
mean_tnr = scr["test_tnr"].mean()
time_i = scr["fit_time"].mean()
print(f"Default SVC:")
print(f"  Mean Balanced Accuracy = {mean_balanced_accuracy}")
print(f"  Mean Accuracy          = {mean_accuracy}")
print(f"  Mean TPR               = {mean_tpr}")
print(f"  Mean TNR               = {mean_tnr}")
print(f"  Training Time          = {time_i:.4f} sec\n")

svc_grid = {
    "classifier__C": [ 1, 10],
    "classifier__kernel": ["linear", "rbf", "poly", "sigmoid"],
    "classifier__gamma": ["scale", "auto"],
    "classifier__degree": [3, 4, 5],  # Solo para kernel "poly"
    "classifier__coef0": [0.0, 1.0]  # Solo para kernel "poly" y "sigmoid"
}

grid = GridSearchCV(clf, svc_grid, cv=cv, scoring=scoring, refit='balanced_accuracy', n_jobs=-1) # Busqueda de hiperparámetros

grid.fit(X_train, y_train)
best_model_svc = grid.best_estimator_ # Mejor modelo encontrado

start_time = time.time()
best_model_svc.fit(X_train, y_train)  # Reentrenar solo con los mejores hiperparámetros
best_fit_time = time.time() - start_time
# Extraer resultados
cv_results = grid.cv_results_

mean_acc = np.max(cv_results["mean_test_accuracy"])  # Mejor Accuracy
mean_tpr = np.max(cv_results["mean_test_tpr"])  # Mejor TPR
mean_tnr = np.max(cv_results["mean_test_tnr"])  # Mejor TNR

print(f"SVC with HPO:")
print(f"  Mean Balanced Accuracy = {grid.best_score_}")
print(f"  Mean Accuracy          = {mean_acc}")
print(f"  Mean TPR               = {mean_tpr}")
print(f"  Mean TNR               = {mean_tnr}")
print(f"  Training Time          = {best_fit_time:.4f} sec\n")
print("Mejores hiperparámetros:", grid.best_params_)
```

## __*7. Tarea de elección abierta: RandomForest*__

Hemos decidido elegir el RandomForest ya que es considerado uno de los algoritmos más certeros del momento, es robusto al overfitting, es capáz de manejar eficientemente bases de datos extensas con muchos atributos (es escalable) y porque nos proporciona estimaciones de que variables son más importantes a la hora de clasificar. Para evaluar el modelo RandomForest, vamos a seguir el mismo procedimiento que en el resto de los modelos.

```python
from sklearn import ensemble


clf = Pipeline([("preprocessor", preprocessor), ("classifier", ensemble.RandomForestClassifier(random_state=seed))]) # Definir el clasificador RandomForest en el pipeline

scr = cross_validate(clf, X_train, y_train, scoring=scoring, cv=cv) # evaluacion sin HPO

# Metricas obtenidas del cross-validation
mean_balanced_accuracy = scr["test_balanced_accuracy"].mean()
mean_accuracy = scr["test_accuracy"].mean()
mean_tpr = scr["test_tpr"].mean()
mean_tnr = scr["test_tnr"].mean()
time_i = scr["fit_time"].mean()
print(f"RandomForest sin HPO:")
print(f"  Mean Balanced Accuracy = {mean_balanced_accuracy}")
print(f"  Mean Accuracy          = {mean_accuracy}")
print(f"  Mean TPR               = {mean_tpr}")
print(f"  Mean TNR               = {mean_tnr}")
print(f"  Training Time          = {time_i:.4f} sec\n")

rf_grid = {
    'classifier__criterion': ['gini', 'entropy'],  # Función de evaluación
    'classifier__max_depth': [2, 5, 10, None],  # Profundidad máxima del árbol
    'classifier__min_samples_split': [2, 10, 50],  # Mínimo de muestras para dividir un nodo
    'classifier__min_impurity_decrease': [0.0, 0.01, 0.05]  # Umbral para crear nuevas divisiones
}

grid = GridSearchCV(clf, rf_grid, cv=cv, scoring=scoring, refit='balanced_accuracy', n_jobs=-1) # Busqueda de los mejores hiperparámetros

grid.fit(X_train, y_train)
best_model_rf = grid.best_estimator_

start_time = time.time()
best_model_rf.fit(X_train, y_train)  # Reentrenar solo con los mejores hiperparámetros
best_fit_time = time.time() - start_time
# Extraer resultados
cv_results = grid.cv_results_

mean_acc = np.max(cv_results["mean_test_accuracy"])  # Mejor Accuracy
mean_tpr = np.max(cv_results["mean_test_tpr"])  # Mejor TPR
mean_tnr = np.max(cv_results["mean_test_tnr"])  # Mejor TNR

print(f"RandomForest con HPO:")
print(f"  Mean Balanced Accuracy = {grid.best_score_}")
print(f"  Mean Accuracy          = {mean_acc}")
print(f"  Mean TPR               = {mean_tpr}")
print(f"  Mean TNR               = {mean_tnr}")
print(f"  Training Time          = {best_fit_time:.4f} sec\n")
print("Mejores hiperparámetros:", grid.best_params_)
```

Se obtienen mejores métricas que con el Decision Tree realizado anteriormente, pero es peor que el SVM por lo que no modificaremos el modelo final. Los hiperparámetros optimizados son los mismos que se obtuvieron en el arbol de decisión (ya que este método se basa en los arboles de decisión básicos)

Como el modelo RandomForest crea varios arboles de decisión con atributos aleatorios, se puede obtener la importancia de los diversos atributos inciales sobre el modelo creado, de la misma manera que se ha hecho en el apartado 5.3 con la regresión logística. 

```python
importancias = best_model_rf.named_steps["classifier"].feature_importances_

# Acceder al OneHotEncoder
ohe = best_model_rf.named_steps["preprocessor"] \
            .named_transformers_["cat"] \
            .named_steps["encoder"]

# Obtener nombres de variables transformadas
ohe_feature_names = ohe.get_feature_names_out(columnas_cat)

# Combinar nombres finales
final_feature_names = list(columnas_num) + list(ohe_feature_names)

# Crear DataFrame con importancias
feature_importance = pd.DataFrame({
    'Característica': final_feature_names,
    'Importancia': importancias
})

# Ordenar por importancia
importancia_agrupada = feature_importance.sort_values(by="Importancia", ascending=False)

print(importancia_agrupada)
```
